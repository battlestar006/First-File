```python
# python function Arguments
def add_numbers(a, b):
    sum = a + b
    print('Sum Is :', sum)

add_numbers(10, 11)

```

```python
#Function Argument with Default Values
def add_numbers( a = 7,  b = 8):
    sum = a + b
    print('Sum:', sum)


# function two arguments
add_numbers(1,1)

add_numbers(a = 2)

add_numbers()
```

```python
#Python Keyword Argument
def display_info(first_name, last_name):
    print('First Name:', first_name)
    print('Last Name:', last_name)

display_info(last_name = 'Jagtap', first_name = 'Sujit')
```

```python
#Python Function With Arbitrary(Variable) Arguments
# program to find sum of multiple numbers

def find_sum(*numbers):#We use an asterisk (*) before the parameter name to denote this kind of argument
    result = 0

    for num in numbers:
        result = result + num

    print("Sum = ", result)

# function call with 3 arguments
find_sum(1, 2, 3)

# function call with 2 arguments
find_sum(4, 9)
```
